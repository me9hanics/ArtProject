```python
import networkx as nx
# This includes the isolated node!
rgo = nx.erdos_renyi_graph(289, 0.01196, seed=20001104, directed=False)
print("Average clustering", nx.average_clustering(rgo))
```

```python
import math

subgraphs = [rgo.subgraph(c).copy() for c in nx.connected_components(rgo) if len(c) > 1]
math.fsum(len(sg) * nx.average_shortest_path_length(sg) for sg in subgraphs) / sum(len(sg) for sg in subgraphs)
```

### Result: InfluencedOn artist network is very "social", it emits a power-law degree distribution, and it has a small-world property.

Table on comparing to a random graph:

| Property | InfluencedOn | Random |
| --- | --- | --- |
| Average path length | 4.7 | 4.49 |
| Average clustering coefficient | 0.126 | 0.023 |

Has about as good path length as a random graph, but multiple times higher clustering coefficient. So we can say that indeed this network is some kind of small-world network.
